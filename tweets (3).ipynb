```
import json
f = open('tweets.json',)
data = json.load(f)
```

**CLASSIFIED THE TWEETS AND THEN GROUPED THEM**

```
import re

count = 0;
def url_match(tweet):
    match = re.match(r'RT\s@....+', tweet['text'])
    if match:
      tweet['tweet_type'] = "RT"
      return "RT"
    else:
        match = re.match(r'@....+', tweet['text'])
        if match:
          tweet['tweet_type'] = "MT"
          return "mention"
        else:
           tweet['tweet_type'] = "TWEET" 
           return "tweet"
```

```
#If i consider only RT
count = 0
for i in data['tweets']:
  flag = url_match(i)
  if(flag == "tweet"):
    count = count +1
count
print('No. of True posts:',count)
```

```
#if we consider MT too
import re

count = 0
def url_match(tweet):
    match = re.match(r'MT\s@....+', tweet['text'])
    match1 = re.match(r'RT\s@....+', tweet['text'])
    match2 = re.match(r'DT\s@....+', tweet['text'])
    match3 = re.match(r'TT\s@....+', tweet['text'])
    if match or match1 or match2 or match3:
      tweet['tweet_type'] = "RT"
      return "RT"
    else:
        match = re.match(r'@....+', tweet['text'])
        if match:
          tweet['tweet_type'] = "MT"
          return "mention"
        else:
           tweet['tweet_type'] = "TWEET" 
           return "tweet" 
```

```

count = 0
for i in data['tweets']:
  flag = url_match(i)
  if(flag == "tweet"):
    count = count +1
count
print('No. of True posts:',count)
```

```
data = sorted(data['tweets'], key = lambda x: x['created_at'])
```

```
only_tweets =[]
for i in data:
  if(i['tweet_type']=='TWEET'):
    only_tweets.append(i)
```

```
groups1 = []
temp = 0
for i in only_tweets:
  twtxt = i['text']
  temp_grp = []
  temp_grp.append(i)
  for j in data:
    if j['tweet_type'] != 'TWEET' :
      gtw = j['text']
      l2 = len(gtw)//2
      if gtw[50:100] in twtxt or gtw[-50:]in twtxt or gtw[l2-20:l2+20] in twtxt:
        temp_grp.append(j)
  groups1.append(temp_grp)
```

```
newgroups = sorted(groups1, key=len, reverse=True)
ranks = 1
for group in newgroups:
    for tweet in group:
        if(tweet['tweet_type']=='TWEET'):
          tweet['Rank'] = ranks
          ranks = ranks + 1
        print(tweet, end='\n')
    print("\n\n")
```
